## 1. Import library

Di bagian ini kita menyiapkan alat gambar dan hitung.
- `numpy` untuk matematika
- `matplotlib` untuk gambar
- `FuncAnimation` untuk simulasi bergerak


```python
import math
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Circle, Rectangle, Polygon, Arc
from matplotlib.animation import FuncAnimation
from IPython.display import HTML, display

plt.rcParams["figure.figsize"] = (8, 5)
plt.rcParams["axes.grid"] = True
plt.rcParams["grid.alpha"] = 0.25
plt.rcParams["font.size"] = 11
```

## 2. Bentuk dasar

Kita mulai dari bangun yang paling mudah dilihat dulu. Tujuannya bukan hanya hafal namanya, tetapi juga melihat sisi, sudut, dan ukuran dasarnya.


```python
fig, axes = plt.subplots(1, 4, figsize=(16, 4))

# Lingkaran
ax = axes[0]
ax.set_title("Lingkaran")
circle = Circle((0, 0), 1, fill=False, linewidth=2, color="#2563EB")
ax.add_patch(circle)
ax.plot([0, 1], [0, 0], color="#22C55E", linewidth=2)
ax.text(0.45, 0.1, "r = 1", color="#22C55E")
ax.set_aspect("equal")
ax.set_xlim(-1.4, 1.4)
ax.set_ylim(-1.4, 1.4)

# Persegi
ax = axes[1]
ax.set_title("Persegi")
square = Rectangle((-1, -1), 2, 2, fill=False, linewidth=2, color="#2563EB")
ax.add_patch(square)
ax.text(-0.15, 0.05, "sisi = 2", color="#0F172A")
ax.set_aspect("equal")
ax.set_xlim(-1.6, 1.6)
ax.set_ylim(-1.6, 1.6)

# Segitiga
ax = axes[2]
ax.set_title("Segitiga")
triangle_points = np.array([[-1, -1], [1, -1], [0, 1]])
triangle = Polygon(triangle_points, closed=True, fill=False, linewidth=2, color="#2563EB")
ax.add_patch(triangle)
ax.text(-0.95, -1.2, "alas")
ax.text(0.1, 0.1, "tinggi")
ax.set_aspect("equal")
ax.set_xlim(-1.6, 1.6)
ax.set_ylim(-1.6, 1.6)

# Persegi panjang
ax = axes[3]
ax.set_title("Persegi panjang")
rect = Rectangle((-1.5, -0.8), 3.0, 1.6, fill=False, linewidth=2, color="#2563EB")
ax.add_patch(rect)
ax.text(-0.2, 0.1, "p x l")
ax.set_aspect("equal")
ax.set_xlim(-2, 2)
ax.set_ylim(-1.5, 1.5)

for ax in axes:
    ax.axhline(0, color="#94A3B8", linewidth=0.8)
    ax.axvline(0, color="#94A3B8", linewidth=0.8)

plt.tight_layout()
plt.show()
```

## 3. Sudut dan jenis segitiga

Sudut itu seperti bukaan pintu. Kalau bukaan kecil, sudutnya kecil. Kalau bukaan lebar, sudutnya besar.


```python
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# Segitiga siku-siku
ax = axes[0]
ax.set_title("Segitiga siku-siku")
right_triangle = np.array([[0, 0], [4, 0], [0, 3]])
ax.add_patch(Polygon(right_triangle, closed=True, fill=False, linewidth=2, color="#2563EB"))
ax.plot([0, 0.4, 0.4, 0], [0, 0, 0.4, 0.4], color="#22C55E")
ax.text(1.7, -0.3, "alas = 4")
ax.text(-0.6, 1.5, "tinggi = 3")
ax.text(1.5, 1.2, "siku-siku")
ax.set_aspect("equal")
ax.set_xlim(-0.8, 4.5)
ax.set_ylim(-0.8, 3.8)

# Segitiga sama kaki
ax = axes[1]
ax.set_title("Segitiga sama kaki")
isosceles = np.array([[0, 0], [4, 0], [2, 3]])
ax.add_patch(Polygon(isosceles, closed=True, fill=False, linewidth=2, color="#2563EB"))
ax.text(1.8, -0.3, "dua sisi sama")
ax.text(0.9, 1.6, "sudut kiri = kanan")
ax.set_aspect("equal")
ax.set_xlim(-0.8, 4.8)
ax.set_ylim(-0.8, 3.8)

# Segitiga sama sisi
ax = axes[2]
ax.set_title("Segitiga sama sisi")
equilateral = np.array([[0, 0], [4, 0], [2, 3.46]])
ax.add_patch(Polygon(equilateral, closed=True, fill=False, linewidth=2, color="#2563EB"))
ax.text(1.3, -0.3, "semua sisi sama")
ax.text(1.5, 1.6, "semua sudut 60°")
ax.set_aspect("equal")
ax.set_xlim(-0.8, 4.8)
ax.set_ylim(-0.8, 4.2)

for ax in axes:
    ax.axhline(0, color="#94A3B8", linewidth=0.8)
    ax.axvline(0, color="#94A3B8", linewidth=0.8)
    ax.grid(False)

plt.tight_layout()
plt.show()
```

## 4. Sin dan cos

Sin dan cos membantu kita melihat posisi titik pada lingkaran. Kalau sudut berubah, posisi titik ikut berubah.


```python
angle_deg = 45
angle_rad = np.deg2rad(angle_deg)

fig, axes = plt.subplots(1, 3, figsize=(16, 4))

# Lingkaran satuan
ax = axes[0]
unit_circle = Circle((0, 0), 1, fill=False, linewidth=2, color="#2563EB")
ax.add_patch(unit_circle)
ax.plot([0, np.cos(angle_rad)], [0, np.sin(angle_rad)], color="#22C55E", linewidth=2)
ax.scatter([np.cos(angle_rad)], [np.sin(angle_rad)], color="#0F172A", s=40)
ax.plot([0, np.cos(angle_rad)], [np.sin(angle_rad), np.sin(angle_rad)], color="#F59E0B", linestyle="--")
ax.plot([np.cos(angle_rad), np.cos(angle_rad)], [0, np.sin(angle_rad)], color="#F59E0B", linestyle="--")
ax.text(0.08, 0.08, "(cos θ, sin θ)")
ax.set_title("Lingkaran satuan")
ax.set_aspect("equal")
ax.set_xlim(-1.3, 1.3)
ax.set_ylim(-1.3, 1.3)

# Gelombang sin
x = np.linspace(0, 2 * np.pi, 400)
axes[1].plot(x, np.sin(x), color="#2563EB", linewidth=2)
axes[1].axvline(angle_rad, color="#22C55E", linestyle="--")
axes[1].axhline(np.sin(angle_rad), color="#22C55E", linestyle="--")
axes[1].scatter([angle_rad], [np.sin(angle_rad)], color="#0F172A")
axes[1].set_title("Grafik sin")
axes[1].set_xlabel("Sudut (radian)")
axes[1].set_ylabel("sin(θ)")
axes[1].set_ylim(-1.2, 1.2)

# Gelombang cos
axes[2].plot(x, np.cos(x), color="#2563EB", linewidth=2)
axes[2].axvline(angle_rad, color="#22C55E", linestyle="--")
axes[2].axhline(np.cos(angle_rad), color="#22C55E", linestyle="--")
axes[2].scatter([angle_rad], [np.cos(angle_rad)], color="#0F172A")
axes[2].set_title("Grafik cos")
axes[2].set_xlabel("Sudut (radian)")
axes[2].set_ylabel("cos(θ)")
axes[2].set_ylim(-1.2, 1.2)

plt.tight_layout()
plt.show()

print(f"Sudut = {angle_deg}°")
print(f"sin({angle_deg}°) = {np.sin(angle_rad):.3f}")
print(f"cos({angle_deg}°) = {np.cos(angle_rad):.3f}")
```

## 5. Simulasi gerak melingkar

Gerak melingkar itu penting karena titik di lingkaran berubah terus. Dari sini kita bisa lihat kenapa sin dan cos muncul di banyak tempat.


```python
def animate_circle_motion(frames=120):
    fig, ax = plt.subplots(figsize=(6, 6))
    ax.set_aspect("equal")
    ax.set_xlim(-1.4, 1.4)
    ax.set_ylim(-1.4, 1.4)
    ax.axhline(0, color="#94A3B8", linewidth=0.8)
    ax.axvline(0, color="#94A3B8", linewidth=0.8)
    ax.add_patch(Circle((0, 0), 1, fill=False, linewidth=2, color="#2563EB"))

    point, = ax.plot([], [], "o", color="#0F172A", markersize=8)
    radius_line, = ax.plot([], [], color="#22C55E", linewidth=2)
    x_proj, = ax.plot([], [], color="#F59E0B", linestyle="--")
    y_proj, = ax.plot([], [], color="#F59E0B", linestyle="--")
    trail_x, trail_y = [], []
    trail, = ax.plot([], [], color="#2563EB", alpha=0.35)
    label = ax.text(-1.25, 1.15, "", fontsize=12)

    def init():
        point.set_data([], [])
        radius_line.set_data([], [])
        x_proj.set_data([], [])
        y_proj.set_data([], [])
        trail.set_data([], [])
        label.set_text("")
        return point, radius_line, x_proj, y_proj, trail, label

    def update(frame):
        theta = 2 * np.pi * frame / frames
        x = np.cos(theta)
        y = np.sin(theta)
        trail_x.append(x)
        trail_y.append(y)
        point.set_data([x], [y])
        radius_line.set_data([0, x], [0, y])
        x_proj.set_data([0, x], [y, y])
        y_proj.set_data([x, x], [0, y])
        trail.set_data(trail_x, trail_y)
        label.set_text(f"θ = {np.rad2deg(theta):.0f}°   x = cos θ   y = sin θ")
        return point, radius_line, x_proj, y_proj, trail, label

    anim = FuncAnimation(fig, update, frames=frames, init_func=init, interval=60, blit=True)
    plt.close(fig)
    return HTML(anim.to_jshtml())

animate_circle_motion()
```

## 6. Segitiga siku-siku dan Pythagoras

Kalau kita tahu dua sisi segitiga siku-siku, kita bisa mencari sisi miring dengan rumus Pythagoras:

$$a^2 + b^2 = c^2$$


```python
a = 3
b = 4
c = math.sqrt(a**2 + b**2)

fig, ax = plt.subplots(figsize=(7, 5))
triangle = np.array([[0, 0], [a, 0], [0, b]])
ax.add_patch(Polygon(triangle, closed=True, fill=False, linewidth=2, color="#2563EB"))
ax.plot([0, 0.35, 0.35, 0], [0, 0, 0.35, 0.35], color="#22C55E")
ax.text(a / 2, -0.35, f"a = {a}")
ax.text(-0.55, b / 2, f"b = {b}")
ax.text(a / 2, b / 2, f"c = {c:.1f}")
ax.text(0.4, 1.2, f"{a}² + {b}² = {a**2} + {b**2} = {c**2:.0f}")
ax.set_aspect("equal")
ax.set_xlim(-1, 5)
ax.set_ylim(-1, 5)
ax.set_title("Pythagoras pada segitiga siku-siku")
plt.show()

print(f"Sisi miring c = {c:.2f}")
```

## 7. Fungsi trigonometri yang bisa kita ubah-ubah

Di bagian ini, kita lihat apa yang terjadi kalau sudutnya diganti. Bayangkan kita memutar titik pada lingkaran satuan.


```python
def show_trig_snapshot(angle_deg):
    angle_rad = np.deg2rad(angle_deg)
    sin_value = np.sin(angle_rad)
    cos_value = np.cos(angle_rad)
    tan_value = np.tan(angle_rad)

    fig, axes = plt.subplots(1, 2, figsize=(13, 5))

    # Lingkaran satuan
    ax = axes[0]
    ax.add_patch(Circle((0, 0), 1, fill=False, linewidth=2, color="#2563EB"))
    ax.plot([0, cos_value], [0, sin_value], color="#22C55E", linewidth=2)
    ax.scatter([cos_value], [sin_value], color="#0F172A", s=50)
    ax.plot([0, cos_value], [sin_value, sin_value], color="#F59E0B", linestyle="--")
    ax.plot([cos_value, cos_value], [0, sin_value], color="#F59E0B", linestyle="--")
    ax.set_title(f"Sudut {angle_deg}°")
    ax.text(-1.15, 1.1, f"sin = {sin_value:.2f}")
    ax.text(-1.15, 0.9, f"cos = {cos_value:.2f}")
    ax.text(-1.15, 0.7, f"tan = {tan_value:.2f}" if abs(cos_value) > 1e-8 else "tan tidak terdefinisi")
    ax.set_aspect("equal")
    ax.set_xlim(-1.4, 1.4)
    ax.set_ylim(-1.4, 1.4)
    ax.axhline(0, color="#94A3B8", linewidth=0.8)
    ax.axvline(0, color="#94A3B8", linewidth=0.8)

    # Bar nilai
    labels = ["sin", "cos", "tan"]
    values = [sin_value, cos_value, tan_value if abs(cos_value) > 1e-8 else 0]
    axes[1].bar(labels, values, color=["#2563EB", "#22C55E", "#F59E0B"])
    axes[1].axhline(0, color="#0F172A", linewidth=0.8)
    axes[1].set_ylim(-2, 2)
    axes[1].set_title("Nilai fungsi")

    plt.tight_layout()
    plt.show()
    print(f"Sudut = {angle_deg}°")
    print(f"sin({angle_deg}°) = {sin_value:.3f}")
    print(f"cos({angle_deg}°) = {cos_value:.3f}")
    if abs(cos_value) > 1e-8:
        print(f"tan({angle_deg}°) = {tan_value:.3f}")
    else:
        print("tan tidak terdefinisi untuk sudut ini")

show_trig_snapshot(30)
```

## 8. Contoh di dunia nyata

Trigonometri bukan cuma untuk soal sekolah. Ia juga dipakai saat menghitung gerak ayunan, jalur orbit, dan tinggi bangunan.


```python
# 1) Ayunan sederhana
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

t = np.linspace(0, 2 * np.pi, 400)
angles = np.deg2rad(30) * np.cos(t)

axes[0].plot(t, np.rad2deg(angles), color="#2563EB")
axes[0].set_title("Ayunan: sudut berubah terhadap waktu")
axes[0].set_xlabel("waktu")
axes[0].set_ylabel("sudut (derajat)")

axes[1].plot(np.cos(angles), np.sin(angles), color="#22C55E")
axes[1].add_patch(Circle((0, 0), 1, fill=False, color="#2563EB"))
axes[1].set_aspect("equal")
axes[1].set_title("Lintasan di lingkaran")
axes[1].set_xlim(-1.2, 1.2)
axes[1].set_ylim(-1.2, 1.2)

# 2) Orbit sederhana
orbit_theta = np.linspace(0, 2 * np.pi, 200)
axes[2].plot(1.6 * np.cos(orbit_theta), 0.9 * np.sin(orbit_theta), color="#F59E0B")
axes[2].scatter([1.6], [0], color="#0F172A")
axes[2].set_aspect("equal")
axes[2].set_title("Orbit elips sederhana")
axes[2].set_xlim(-2.2, 2.2)
axes[2].set_ylim(-1.5, 1.5)

plt.tight_layout()
plt.show()

# 3) Mengukur tinggi bangunan dengan bayangan
jarak_bayangan = 12
sudut_matahari = 35
tinggi = jarak_bayangan * math.tan(math.radians(sudut_matahari))
print(f"Jika jarak bayangan = {jarak_bayangan} m dan sudut matahari = {sudut_matahari}°, maka tinggi = {tinggi:.2f} m")
```
